```python
import pandas as pd
import xgboost as xgb
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data_forex.csv')
```

```python
data['High'].plot()
```

```python
train_data = data.iloc[:int(.99*len(data)), :]
test_data = data.iloc[int(.99*len(data)):, :]
```

```python
features = ['Open', 'High', 'Low', 'Close']
target = 'Close'
```

```python
model = xgb.XGBRegressor()
model.fit(train_data[features], train_data[target])
```

```python
predictions = model.predict(test_data[features])
print('Model Predictions:')
print(predictions)
```

```python
print('Actual Values:')
print(test_data[target])
```

```python
accuracy = model.score(test_data[features], test_data[target])
print('Accuracy:')
print(accuracy)
```

```python
plt.plot(test_data['Close'], label = 'Close Price')
plt.plot(test_data[target].index, predictions, label = 'Predictions')
plt.legend()
plt.show()
```
